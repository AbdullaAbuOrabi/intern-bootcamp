```python
orders["order_date"] = pd.to_datetime(orders["order_date"])

sales = (
    order_items
    .merge(orders, on="order_id", how="left")
    .merge(products, on="product_id", how="left")
    .merge(customers, on="customer_id", how="left")
)

sales.head()
```

```python
daily_revenue = (
    sales
    .groupby(sales["order_date"].dt.date)["total_price"]
    .sum()
    .reset_index()
)

daily_revenue.columns = ["order_date", "revenue"]

daily_revenue.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(daily_revenue["order_date"], daily_revenue["revenue"], marker="o")
plt.title("Revenue Over Time")
plt.xlabel("Order Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_customers = (
    sales
    .groupby("customer_id")["total_price"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_customers
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_customers, x="total_price", y="customer_id")
plt.title("Top 10 Customers by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Customer ID")
plt.tight_layout()
plt.show()
```

```python
category_revenue = (
    sales
    .groupby("category")["total_price"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

category_revenue
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=category_revenue, x="total_price", y="category")
plt.title("Revenue by Product Category")
plt.xlabel("Total Revenue")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=sales, x="total_price")
plt.title("Boxplot of Order Item Total Price")
plt.xlabel("Total Price")
plt.tight_layout()
plt.show()
```

```python
numeric_columns = sales[["quantity", "unit_price", "total_price"]]

correlation_matrix = numeric_columns.corr()

correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="Blues")
plt.title("Correlation Heatmap for Numeric Variables")
plt.tight_layout()
plt.show()
```

## Key Findings

### 1. Revenue Over Time
The revenue over time chart shows how total sales changed across different order dates. Some days had higher revenue than others, which can help identify peak sales periods.

### 2. Top Customers by Revenue
The top customers chart shows that some customers generated more revenue than others. This helps identify the most valuable customers for the business.

### 3. Product Category Performance
The revenue by product category chart shows which product categories contributed the most to total revenue. This can help the business understand which categories are performing best.

### 4. Outlier Detection
The boxplot helped identify possible outliers in total order item prices. Outliers are values that are much higher or lower than most other values.

### 5. Correlation Analysis
The correlation heatmap showed the relationship between quantity, unit price, and total price. This helps understand how numeric variables are connected.
