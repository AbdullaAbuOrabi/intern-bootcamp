# MongoDB Fundamentals

## Objective

The objective of this notebook is to understand MongoDB basics, create collections, insert documents, and query documents using Python and pymongo.

MongoDB is a NoSQL database that stores data as flexible documents instead of fixed tables.

## SQL vs MongoDB Terms

| PostgreSQL / SQL | MongoDB |
|---|---|
| Database | Database |
| Table | Collection |
| Row | Document |
| Column | Field |
| SQL query | MongoDB query |

In PostgreSQL, data is stored in tables with rows and columns.

In MongoDB, data is stored in collections as documents. A document looks similar to a JSON object or Python dictionary.
