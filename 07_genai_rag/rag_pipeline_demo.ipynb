# RAG Pipeline Demo

This notebook demonstrates a simple Retrieval-Augmented Generation pipeline.

The pipeline follows these steps:

1. Retrieve relevant context from the vector database.
2. Build a prompt using the retrieved context.
3. Generate a direct answer.
4. Log the query, answer, context relevance, and response quality.

```python
import os

os.chdir(r"C:\Users\homam\OneDrive\Desktop\intern-bootcamp\07_genai_rag")

print("Current folder:", os.getcwd())
```

```python
from rag_pipeline import rag_answer
```

```python
test_queries = [
    "What is semantic search?",
    "What is a vector store?",
    "Why do we use chunking?",
    "What does a vector store save?",
    "How does semantic search find text?"
]
```

```python
results = []

for query in test_queries:
    result = rag_answer(query)
    results.append(result)

    print("=" * 80)
    print("QUESTION:")
    print(result["query"])

    print("\nANSWER:")
    print(result["answer"])

    print("\nCONTEXT RELEVANCE:")
    print(result["context_relevance"])

    print("\nRESPONSE QUALITY:")
    print(result["response_quality"])

    print("\nSTATUS:")
    print(result["status"])
    print()
```

```python
from rag_pipeline import save_results

save_results(results, "rag_results.json")

print("Results saved to rag_results.json")
```
