# Day 4 — Evaluating and Refining LLM Outputs

## Objective

The goal of this task is to evaluate the quality of LLM outputs using simple manual and rule-based evaluation methods.

Instead of only generating answers from an LLM, this task focuses on checking whether the answers are faithful, coherent, complete, concise, and relevant.

We will compare multiple prompt versions for the same product review and score their outputs using Python.

```python
review = """
The wireless headphones are very comfortable and the sound quality is great,
but the battery drains faster than expected.
"""

print(review)
```

## Prompt Versions

We will test multiple prompt styles to compare how the LLM responds.

The prompt versions are:

1. Vague prompt
2. Structured prompt
3. Few-shot prompt
4. Overly constrained prompt

```python
prompts = [
    {
        "prompt_version": "Prompt 1 - Vague",
        "prompt": "Summarize this product review."
    },
    {
        "prompt_version": "Prompt 2 - Structured",
        "prompt": """
Analyze the following product review.

Return the answer using this format:
- Sentiment:
- Positive points:
- Negative points:
- Recommendation:
"""
    },
    {
        "prompt_version": "Prompt 3 - Few-shot",
        "prompt": """
You are a product review analyst.

Example:
Review:
The phone has a great camera but poor battery life.

Answer:
- Sentiment: Mixed
- Positive points: Great camera
- Negative points: Poor battery life
- Recommendation: Good for users who care about camera quality, but not ideal for users who need long battery life.

Now analyze the given review using the same format.
"""
    },
    {
        "prompt_version": "Prompt 4 - Overly Constrained",
        "prompt": "Summarize the review in exactly one short sentence."
    }
]

for item in prompts:
    print(item["prompt_version"])
    print(item["prompt"])
    print("-" * 50)
```

```python
outputs = [
    {
        "prompt_version": "Prompt 1 - Vague",
        "output": "The headphones are comfortable and have good sound quality, but the battery life is disappointing."
    },
    {
        "prompt_version": "Prompt 2 - Structured",
        "output": """
- Sentiment: Mixed
- Positive points: Comfortable design, great sound quality
- Negative points: Battery drains faster than expected
- Recommendation: Good for users who care about comfort and sound quality, but not ideal for users who need long battery life.
"""
    },
    {
        "prompt_version": "Prompt 3 - Few-shot",
        "output": """
- Sentiment: Mixed
- Positive points: Very comfortable, great sound quality
- Negative points: Battery drains faster than expected
- Recommendation: Good for users who value comfort and audio quality, but not recommended for users who need strong battery performance.
"""
    },
    {
        "prompt_version": "Prompt 4 - Overly Constrained",
        "output": "Comfortable headphones with great sound, but weak battery life."
    }
]
```
