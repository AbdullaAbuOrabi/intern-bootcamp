# Document Chunking and Vector Store Creation

This notebook demonstrates how to load a markdown document, split it into smaller chunks, add metadata, create embeddings, store the chunks in a Chroma vector store, and retrieve relevant chunks using semantic search.


## 1. Load the Markdown Document

In this step, we load the project FAQ markdown file from the `source_documents` folder.
This document will be used as the source data for chunking and vector store creation.

```python
from pathlib import Path

# Path to the markdown document
document_path = Path("../source_documents/faq.md")

# Read the file content
text = document_path.read_text(encoding="utf-8")

print("Document loaded successfully")
print("--------------------------------")
print(text)
```

## 2. Split the Document into Chunks

In this step, we use LangChain's `RecursiveCharacterTextSplitter` to split the markdown document into smaller chunks.

Chunking helps the retrieval system search smaller and more focused pieces of text instead of searching one large document.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Create the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=40
)

# Split the document into chunks
chunks = text_splitter.split_text(text)

print(f"Number of chunks created: {len(chunks)}")
print("--------------------------------")

for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}:")
    print(chunk)
    print("--------------------------------")
```

## 3. Add Metadata to Each Chunk

In this step, each text chunk is converted into a LangChain `Document` object.

Each document contains the chunk text and metadata such as the source file name, chunk number, and document type.

```python
from langchain_core.documents import Document

documents = []

for i, chunk in enumerate(chunks, start=1):
    document = Document(
        page_content=chunk,
        metadata={
            "source": "faq.md",
            "chunk_number": i,
            "document_type": "FAQ"
        }
    )
    documents.append(document)

print(f"Number of Document objects created: {len(documents)}")
print("--------------------------------")

for doc in documents:
    print("Chunk text:")
    print(doc.page_content)
    print("Metadata:")
    print(doc.metadata)
    print("--------------------------------")
```

## 4. Create Embeddings and Store in Chroma

In this step, we create embeddings for each chunk using a HuggingFace embedding model.

The embeddings and chunks are stored in a Chroma vector store under the `vector_store` folder.

```python
%pip install langchain-chroma langchain-huggingface sentence-transformers chromadb
```

```python
from pathlib import Path
import shutil

from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

# Path where the vector store will be saved
vector_store_path = Path("../vector_store")

# Clear old vector store if it already exists
if vector_store_path.exists():
    shutil.rmtree(vector_store_path)

# Create embedding model
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embedding model loaded successfully")
print("--------------------------------")

# Create and save Chroma vector store
vector_store = Chroma.from_documents(
    documents=documents,
    embedding=embedding_model,
    persist_directory=str(vector_store_path),
    collection_name="project_faq"
)

print("Vector store created successfully")
print(f"Vector store saved at: {vector_store_path}")
```

## 5. Test Semantic Retrieval

In this step, we query the Chroma vector store to retrieve the most relevant chunks for a sample question.

This confirms that the stored embeddings can be searched by meaning.

```python
query = "Why do we split documents into chunks?"

results = vector_store.similarity_search(query, k=3)

print(f"Query: {query}")
print("--------------------------------")

for i, result in enumerate(results, start=1):
    print(f"Result {i}:")
    print(result.page_content)
    print("Metadata:")
    print(result.metadata)
    print("--------------------------------")
```
