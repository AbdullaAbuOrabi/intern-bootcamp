# Day 2 — Principles of Prompt Design

## Task Goal

The goal of this task is to understand how different prompt styles affect the output of a Large Language Model.

In this notebook, I will test different prompt formats:

1. Zero-shot prompting
2. One-shot prompting
3. Few-shot prompting
4. Reasoning-style prompting

I will use the same product review task and compare how the output changes depending on the prompt design.

## Why this task is important

Prompts are the instructions given to an LLM. A clear and structured prompt helps the model produce better, more useful, and more consistent answers.

This task helps me understand how to write better prompts for future GenAI projects.

## Sample Product Reviews

For this task, I will use product reviews as the example use case.

The LLM will be asked to summarize product reviews and identify:

- Customer sentiment
- Main positive points
- Main negative points
- Final recommendation

```python
reviews = [
    "The wireless headphones are very comfortable and the sound quality is great, but the battery drains faster than expected.",
    "The laptop is fast, lightweight, and easy to carry, but it becomes hot when running heavy programs.",
    "The smartwatch has a beautiful screen and useful fitness tracking features, but the mobile app is confusing to use."
]

for i, review in enumerate(reviews, start=1):
    print(f"Review {i}: {review}")
```

## 1. Zero-Shot Prompting

Zero-shot prompting means asking the LLM to complete a task without giving it any example.

In this section, I will give the model only the instruction and the product review. The model has to understand the task from the prompt alone.

```python
zero_shot_prompt = """
Summarize the following product review.

Review:
The wireless headphones are very comfortable and the sound quality is great, but the battery drains faster than expected.
"""

print(zero_shot_prompt)
```

### Zero-Shot Output

The wireless headphones are comfortable and have great sound quality, but the battery life is disappointing because it drains faster than expected.

### Observation

The zero-shot prompt gave a correct summary. However, the answer was simple and written as one sentence. It did not separate the sentiment, positive points, negative points, or recommendation.

This shows that zero-shot prompting works for simple tasks, but the output may not be structured unless we clearly ask for a specific format.

## 2. One-Shot Prompting

One-shot prompting means giving the LLM one example before asking it to complete the task.

The example helps the model understand the expected format, tone, and structure of the answer.

In this section, I will provide one example of a product review summary, then ask the model to summarize a new review using the same style.

```python
one_shot_prompt = """
You are a product review analyst.

Example:

Review:
The phone has an excellent camera and a bright screen, but the battery life is weak.

Summary:
- Sentiment: Mixed
- Positive points: Excellent camera, bright screen
- Negative points: Weak battery life
- Recommendation: Good for users who care about camera and display quality, but not ideal for users who need long battery life.

Now summarize the following review using the same format:

Review:
The wireless headphones are very comfortable and the sound quality is great, but the battery drains faster than expected.
"""

print(one_shot_prompt)
```

### One-Shot Output

- Sentiment: Mixed
- Positive points: Very comfortable, great sound quality
- Negative points: Battery drains faster than expected
- Recommendation: Good for users who care about comfort and sound quality, but not ideal for users who need long battery life.

### Observation

The one-shot prompt gave a better output than the zero-shot prompt because it followed the example format.

In zero-shot prompting, the model only gave one simple sentence. In one-shot prompting, the model separated the answer into sentiment, positive points, negative points, and recommendation.

This shows that giving one example helps the LLM understand the expected structure and style of the answer.

## 3. Few-Shot Prompting

Few-shot prompting means giving the LLM multiple examples before asking it to complete the task.

The examples help the model understand the expected answer format more clearly.

Compared to one-shot prompting, few-shot prompting can make the output more consistent because the model has more examples to follow.

```python
few_shot_prompt = """
You are a product review analyst.

Example 1:

Review:
The phone has an excellent camera and a bright screen, but the battery life is weak.

Summary:
- Sentiment: Mixed
- Positive points: Excellent camera, bright screen
- Negative points: Weak battery life
- Recommendation: Good for users who care about camera and display quality, but not ideal for users who need long battery life.

Example 2:

Review:
The keyboard feels premium and is comfortable to type on, but it is too expensive.

Summary:
- Sentiment: Mixed
- Positive points: Premium feel, comfortable typing experience
- Negative points: High price
- Recommendation: Good for users who value keyboard quality, but not ideal for budget buyers.

Now summarize the following review using the same format:

Review:
The wireless headphones are very comfortable and the sound quality is great, but the battery drains faster than expected.
"""

print(few_shot_prompt)
```

### Few-Shot Output

- Sentiment: Mixed
- Positive points: Very comfortable, great sound quality
- Negative points: Battery drains faster than expected
- Recommendation: Good for users who value comfort and sound quality, but not ideal for users who need long battery life.

### Observation

The few-shot prompt produced a structured answer using the same format as the examples.

The output was very similar to the one-shot result because the review is simple and the required format is clear. However, few-shot prompting gives the model more examples, which can make the output more consistent when the task is more complex.

This shows that adding more examples helps the LLM understand the expected style, format, and level of detail.

## 4. Reasoning-Style Prompting

Reasoning-style prompting asks the LLM to explain the answer briefly before giving the final result.

This is useful when the task requires analysis or judgment, such as deciding whether a review is positive, negative, or mixed.

In this section, I will ask the model to identify the sentiment, explain the reason briefly, and then summarize the review.

```python
reasoning_prompt = """
You are a product review analyst.

Analyze the following product review.

First, briefly explain why the sentiment is positive, negative, or mixed.
Then provide the final answer using this format:

- Sentiment:
- Brief reasoning:
- Positive points:
- Negative points:
- Recommendation:

Review:
The wireless headphones are very comfortable and the sound quality is great, but the battery drains faster than expected.
"""

print(reasoning_prompt)
```

### Reasoning-Style Output

- Sentiment: Mixed
- Brief reasoning: The review includes strong positive points about comfort and sound quality, but also mentions a clear negative issue with battery life.
- Positive points: Very comfortable, great sound quality
- Negative points: Battery drains faster than expected
- Recommendation: Good for users who care about comfort and sound quality, but not ideal for users who need long battery life.

### Observation

The reasoning-style prompt gave the most explainable answer because it included a brief reason for the sentiment.

Compared to one-shot and few-shot prompting, the output was similar in structure, but it added an explanation showing why the sentiment is mixed.

This type of prompt is useful when the user needs to understand the reason behind the answer, not just the final result.

## Prompt Style Comparison

The table below compares the different prompt styles tested in this notebook.

| Prompt Style | Description | Output Quality | Main Observation |
|---|---|---|---|
| Zero-shot | The model was given only the task instruction without examples. | Basic | The output was correct, but it was very similar to the original review and not structured. |
| One-shot | The model was given one example before the real task. | Better | The output followed the example format and became more organized. |
| Few-shot | The model was given multiple examples before the real task. | Strong | The output was structured and consistent because the model had more examples to follow. |
| Reasoning-style | The model was asked to briefly explain the answer before giving the final result. | Most explainable | The output included a reason for the sentiment, making the answer easier to understand. |

## Final Observation

Changing the prompt changed the quality and structure of the model output.

The zero-shot prompt gave a simple summary, but it did not provide a clear structure.

The one-shot and few-shot prompts improved the output because the examples showed the model how to answer.

The reasoning-style prompt was useful because it explained why the sentiment was mixed.

Overall, this task showed that clear prompts, examples, structure, and constraints help LLMs produce better and more useful responses.
