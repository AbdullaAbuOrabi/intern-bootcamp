# Introduction to Generative AI and LLMs

In this notebook, I test simple text generation using a local LLM through Ollama.  
I will compare different temperature and max token settings and observe how they affect the model output.

```python
import ollama
import pandas as pd

model_name = "llama3.2:1b"

print("Ollama is ready.")
print("Using model:", model_name)
```

## 1. Simple Text Generation

In this section, I send a simple prompt to the local Ollama model and check if it can generate a clear answer.

```python
prompt = "Explain Generative AI in simple words."

response = ollama.generate(
    model=model_name,
    prompt=prompt,
    options={
        "temperature": 0.3,
        "num_predict": 120
    }
)

print(response["response"])
```

## 2. Temperature Experiment

In this section, I test the same prompt using different temperature values.  
Temperature controls how creative or predictable the model output is.

Lower temperature usually gives more direct and stable answers.  
Higher temperature usually gives more creative and varied answers.

```python
prompt = "Explain Generative AI in simple words."

temperatures = [0.1, 0.7, 1.0]

results = []

for temp in temperatures:
    response = ollama.generate(
        model=model_name,
        prompt=prompt,
        options={
            "temperature": temp,
            "num_predict": 120
        }
    )
    
    results.append({
        "Prompt": prompt,
        "Temperature": temp,
        "Max Tokens": 120,
        "Output": response["response"]
    })

temperature_df = pd.DataFrame(results)
temperature_df
pd.set_option("display.max_colwidth", None)
temperature_df
```

## 3. Max Token Experiment

In this section, I test the same prompt with different output length limits.  
The `num_predict` setting controls how long the model response can be.

A smaller value gives a shorter answer.  
A larger value allows the model to generate a longer answer.

```python
prompt = "Explain Large Language Models in simple words."

max_token_values = [50, 120, 250]

token_results = []

for max_tokens in max_token_values:
    response = ollama.generate(
        model=model_name,
        prompt=prompt,
        options={
            "temperature": 0.3,
            "num_predict": max_tokens
        }
    )
    
    token_results.append({
        "Prompt": prompt,
        "Temperature": 0.3,
        "Max Tokens": max_tokens,
        "Output": response["response"]
    })

token_df = pd.DataFrame(token_results)
token_df
```

```python
for index, row in token_df.iterrows():
    print("=" * 80)
    print("Max Tokens:", row["Max Tokens"])
    print("Temperature:", row["Temperature"])
    print("-" * 80)
    print(row["Output"])
    print()
```

## 4. Creative vs Factual Text Generation

In this section, I compare how the model responds to creative prompts and factual prompts.

Creative prompts allow the model to use imagination and storytelling.  
Factual prompts should be more direct, accurate, and clear.

```python
generation_tasks = [
    {
        "Type": "Creative",
        "Prompt": "Write a short creative story about a robot learning Python for the first time.",
        "Temperature": 0.9,
        "Max Tokens": 180
    },
    {
        "Type": "Factual",
        "Prompt": "Explain what a Large Language Model is in simple technical words.",
        "Temperature": 0.2,
        "Max Tokens": 180
    }
]

creative_factual_results = []

for task in generation_tasks:
    response = ollama.generate(
        model=model_name,
        prompt=task["Prompt"],
        options={
            "temperature": task["Temperature"],
            "num_predict": task["Max Tokens"]
        }
    )
    
    creative_factual_results.append({
        "Type": task["Type"],
        "Prompt": task["Prompt"],
        "Temperature": task["Temperature"],
        "Max Tokens": task["Max Tokens"],
        "Output": response["response"]
    })

creative_factual_df = pd.DataFrame(creative_factual_results)
creative_factual_df
```

```python
for index, row in creative_factual_df.iterrows():
    print("=" * 80)
    print("Type:", row["Type"])
    print("Temperature:", row["Temperature"])
    print("Max Tokens:", row["Max Tokens"])
    print("-" * 80)
    print("Prompt:")
    print(row["Prompt"])
    print("-" * 80)
    print("Output:")
    print(row["Output"])
    print()
```

The creative output should sound more like a story, with imagination and expressive language.

The factual output should sound more like an explanation, with a clear and direct answer.
